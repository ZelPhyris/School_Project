## Partie I : Indicateurs statistiques d'une série de données quantitatives

On considère ici des relevés de données numériques (valeurs entières ou réelles) fournis sous la forme de tuples.

Par exemple : 

releves = (2, 1, 4, 3, 5, 2, 10, 4, 6, 3, 2)

```python
def effectif_total(xxx):
    """
    retourne l'effectif, c'est à dire, le nombre de valeurs contenue dans donnees
    :param donnees: tuple ne contenant que des valeurs de type int ou float
    :return: (type int) L'effectif des données
    
    >>> A = (2.61, 10.45, 7.21)
    >>> effectif_total(A)
    3
    
    >>> B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
    >>> effectif_total(B)
    16
    
    """
    
    # code de la fonction
    return len(xxx)

# --- Exercice ---
A = (2.61, 10.45, 7.21)
print(f"Effectif de A : {effectif_total(A)}")

B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
print(f"Effectif de B : {effectif_total(B)}")
```

```python
def somme(xxx):
    """
    retourne la somme des valeurs contenues dans donnees
    :param donnees: tuple ne contenant que des valeurs de type int ou float
    :return: (type int ou float) La somme des valeurs
    
    >>> A = (2.61, 10.45, 7.21)
    >>> somme(A)
    20.27
    
    >>> B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
    >>> somme(B)
    72
    
    """
    
    # code de la fonction
    return sum(xxx)

# --- Exercice ---
A = (2.61, 10.45, 7.21)
print(f"Somme de A : {somme(A)}")

B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
print(f"Somme de B : {somme(B)}")
```

```python
def moyenne(xxx):
    """
    retourne la moyenne des valeurs contenues dans donnees
    :param donnees: tuple ne contenant que des valeurs de type int ou float
    :return: (type float) Moyenne des donnees arrondie à deux décimales
    
    >>> A = (2.61, 10.45, 7.21)
    >>> moyenne(A)
    6.76
    
    >>> B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
    >>> moyenne(B)
    4.5
    
    """
    
    # code de la fonction
    return round(somme(xxx) / effectif_total(xxx), 2)

# --- Exercice ---
A = (2.61, 10.45, 7.21)
print(f"Moyenne de A : {moyenne(A)}")

B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
print(f"Moyenne de B : {moyenne(B)}")
```

```python
def val_max(xxx):
    """
    retourne la valeur maximale des données contenues dans donnees
    :param donnees: tuple ne contenant que des valeurs de type int ou float
    :return: (type int ou float) La valeur maximale dans le tuple
    
    >>> A = (2.61, 10.45, 7.21)
    >>> val_max(A)
    10.45
    
    >>> B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
    >>> val_max(B)
    10
    
    """
    
    # code de la fonction
    return max(xxx)

# --- Exercice ---
A = (2.61, 10.45, 7.21)
print(f"Valeur maximale de A : {val_max(A)}")

B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
print(f"Valeur maximale de B : {val_max(B)}")
```

```python
def val_min(xxx):
    """
    retourne la valeur minimale présente dans donnees
    :param donnees: tuple ne contenant que des valeurs de type int ou float
    :return: (type int ou float) La valeur minimale dans le tuple
    
    >>> A = (2.61, 10.45, 7.21)
    >>> val_min(A)
    2.61
    
    >>> B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
    >>> val_min(B)
    1
    
    """
    
    # code de la fonction
    return min(xxx)

# --- Exercice ---
A = (2.61, 10.45, 7.21)
print(f"Valeur minimale de A : {val_min(A)}")

B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
print(f"Valeur minimale de B : {val_min(B)}")
```

```python
def etendue(xxx):
    """
    retourne l'étendue de la série statistique fournie par donnees

    DEF : L'étendue d'une série statistique est la différence entre la valeur la plus grande et la valeur la plus petite de cette série.

    :param donnees: tuple ne contenant que des valeurs de type int ou float
    :return: (type int ou float) L'étendue de la série de données

    >>> A = (2.61, 10.45, 7.21)
    >>> etendue(A)
    7.84
    
    >>> B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
    >>> etendue(B)
    9
    
    """
    
    # code de la fonction
    return round(val_max(xxx) - val_min(xxx), 2)

# --- Exercice ---
A = (2.61, 10.45, 7.21)
print(f"Étendue de A : {etendue(A)}")

B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
print(f"Étendue de B : {etendue(B)}")
```

```python
def effectif_valeur(xxx, valeur):
    """
    retourne l'effectif de la valeur dans la série de données fournie sous forme de tuple.

    DEF : L'effectif d'une valeur dans un relevé statistique correspond au nombre de fois où cette valeur apparait.

    :param donnees: tuple ne contenant que des valeurs de type int ou float
    :param valeur: (type int ou float) la valeur dont on cherche l'effectif
    :return: (type int) Effectif de la valeur dans la série donnees

    >>> A = (2.61, 10.45, 7.21)
    >>> effectif_valeur(A, 10.45)
    1
    
    >>> B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
    >>> effectif_valeur(B, 10)
    2
    >>> effectif_valeur(B, 5)
    0
    
    """
    
    # code de la fonction
    return xxx.count(valeur)

# --- Exercice ---
A = (2.61, 10.45, 7.21)
print(f"Effectif de 10.45 dans A : {effectif_valeur(A, 10.45)}")

B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
print(f"Effectif de 10 dans B : {effectif_valeur(B, 10)}")
print(f"Effectif de 5 dans B : {effectif_valeur(B, 5)}")
```

```python
def frequence_valeur(xxx, valeur):
    """
    retourne la fréquence de valeur dans la série de données fournie sous la forme
    d'un tuple

    DEF : La fréquence d'une valeur dans un relevé statistique correspond au rapport de l'effectif de cette valeur sur l'effectif total 
    de la série de données.
    
    :param donnees: tuple ne contenant que des valeurs de type int ou float
    :param valeur: (type int ou float) la valeur dont on cherche la fréquence
    : return : (tpe int) La fréquence de valeur dans la série donnees
    
    >>> A = (2.61, 10.45, 7.21)
    >>> frequence_valeur(A, 7.21)
    0.3333333333333333
    
    >>> B=(2,1,3,6,1,2,8,10,3,1,1,3,4,8,9,10)
    >>> frequence_valeur(B,10)
    0.125
    >>> frequence_valeur(B, 5)
    0.0
    
    """
    
    # code de la fonction
    return round(effectif_valeur(xxx, valeur) / effectif_total(xxx), 2)

    # --- Exercice ---
A = (2.61, 10.45, 7.21)
print(f"Fréquence de 7.21 dans A : {frequence_valeur(A, 7.21)}")

B = (2, 1, 3, 6, 1, 2, 8, 10, 3, 1, 1, 3, 4, 8, 9, 10)
print(f"Fréquence de 10 dans B : {frequence_valeur(B, 10)}")
print(f"Fréquence de 5 dans B : {frequence_valeur(B, 5)}")
```

## Partie II : Exploitation des données métérologiques

Dans cette seconde partie, vous allez manipuler des tuples de tuples : chaque valeur du tuple principal étant lui-même un tuple contenant une chaîne de caractères et deux autres tuples.

Dans un tuple encapsulé, la chaîne correspond au nom d’une ville, le premier tuple contient les mesures de température réalisées dans cette ville chaque mois d’une année (il y a donc 12 valeurs, la première correspondant à la température mesurée en janvier), et le second tuple contient des mesures de précipitations réalisées dans cette même ville pour chacun des mois (il y a donc aussi 12 valeurs, la première correspondant aux précipitations mesurées en janvier).

La structure de données manipulée est donc de la forme : 

releves = (("ville_1", (températures_1), (précipitations_1)), …, ("ville_n", (températures_n), (précipitations_n))) 

Ce tuple contient les données météorologiques de huit villes françaises réparties sur la surface de l’hexagone. 

Ces données sont issues du site : https://www.infoclimat.fr 

Ce tuple étant défini en dehors de toute fonction, il correspond à une variable globale accessible de n’importe quel endroit du programme, y compris à l’intérieur des fonctions.

```python
releves = (
    ("Lille",
     (5.9, 7.7, 7.5, 12.9, 14.6, 17.7, 18.6, 21.5, 17.3, 12.0, 9.5, 6.0),
     (24.5, 92.3, 65.7, 41.1, 4.2, 72.2, 25.0, 64.1, 8.6, 109.0, 37.1, 83.3)),
    ("Paris",
     (7.9, 10.1, 9.9, 16.2, 17.1, 19.7, 21.4, 23.6, 20.1, 13.6, 11.5, 7.9),
     (29.6, 105.6, 54.0, 29.4, 47.0, 48.2, 7.8, 34.6, 44.6, 82.8, 14.0, 107.8)),
    ("Strasbourg",
     (4.5, 7.7, 7.9, 13.9, 15.1, 18.9, 21.1, 22.4, 18.0, 11.8, 7.0, 4.8),
     (10.8, 80.0, 47.4, 14.1, 56.0, 95.9, 12.0, 61.9, 18.9, 55.7, 11.4, 43.5)),
    ("Rennes",
     (7.7, 9.4, 9.2, 14.0, 15.8, 16.9, 18.9, 20.1, 17.7, 12.7, 10.7, 7.4),
     (81.5, 66.4, 45.1, 46.6, 59.7, 86.8, 8.2, 48.9, 65.1, 102.8, 23.7, 122.4)),
    ("Bordeaux",
     (9.1, 11.0, 10.6, 15.6, 18.3, 18.7, 22.6, 23.0, 20.1, 14.1, 12.2, 8.6),
     (67.3, 60.7, 103.5, 114.0, 122.0, 91.6, 3.0, 66.8, 90.4, 179.3, 16.9, 241.3)),
    ("Grenoble",
     (4.6, 7.1, 7.7, 12.4, 15.2, 17.8, 21.4, 21.9, 17.8, 11.2, 8.2, 4.4),
     (25.4, 33.0, 60.5, 46.2, 74.3, 94.2, 10.0, 49.5, 86.3, 137.8, 24.7, 112.6)),
    ("Toulouse",
     (8.0, 10.3, 10.7, 15.4, 18.5, 19.8, 23.5, 24.3, 20.4, 13.7, 12.2, 7.6),
     (59.8, 29.7, 57.8, 52.9, 58.3, 88.3, 3.4, 24.6, 38.2, 81.0, 15.1, 116.0)),
    ("Marseille",
     (11.4, 12.3, 12.2, 14.6, 18.8, 20.5, 23.6, 24.1, 20.9, 15.7, 14.8, 10.3),
     (21.2, 4.2, 8.2, 63.2, 67.8, 17.4, 0.2, 4.8, 45.4, 16.6, 19.0, 56.4))
    )
```

```python
def indice_ville(xxx):
    """
    Retourne l'indice du tuple encapsulé contenant ville ou False si ville n'est
    pas présente dans le tuple releves défini en variable globale. 
    :param ville: (type str) Le nom de la ville recherchée
    :return: (int ou bool) Indice du tuple corrrespondnat à la ville
                           False si la ville ne se trouve pas dans releves
    
    >>> indice_ville("Lille")
    0
    
    >>> indice_ville("Grenoble")
    5
    
    >>> indice_ville("Lyon")
    False

    """
    
    # code de la fonction
    return next((i for i, v in enumerate(releves) if v[0] == xxx), False)

    # --- Exercice ---
print(indice_ville("Lille"))
print(indice_ville("Grenoble"))
print(indice_ville("Lyon"))
```

```python
def temp_moyenne(xxx):
    """
    retourne la température moyenne annuelle de la ville arrondie à 2 décimales ;
    les données initiales étant celles du tuple releves.
    :param ville: (type str) Le nom d'une ville
    :return:(type float) La température moyenne annuelle de la ville
    
    :CU: La ville doit être présente dans la variable globale releves
    
    >>> temp_moyenne("Lille")
    12.60
      
    """
    
    # code de la fonction
    return round(moyenne(releves[indice_ville(xxx)][1]), 2)

    # --- Exercice ---
print(temp_moyenne("Lille"))
```

```python
def cumul_precipitations(xxx):
    """
    retourne le cumul annuel des précipitations de la ville, arrondi à 2 décimales ;
    les données initiales étant celles du tuple releves.
    : ville : (type str) Le nom d'une ville
    : return : (type int ou float) Le cumul annuel des précipitations pour cette ville
    
    :CU: La ville doit être présente dans la variable globale releves
    
    >>>cumul_precipitation("Lille")
    627.1
    
    """ 
        
    # code de la fonction
    return round(somme(releves[indice_ville(xxx)][2]), 2)

    # --- Exercice ---
print(cumul_precipitations("Lille"))
```

```python
def valeurs_annuelles_villes():
    """
    Retourne un tuple de tuples, chacun d'eux contenant le nom d'une ville,
    sa température moyenne annuelle et son cumul annuel des précipitations ;
    les données initiales étant celles du tuple releves.
    :param: Aucun
    :return: tuple de tuples contenant chacun le nom d'une ville, sa température moyenne annuelle
             et son cumul annuel des précipitations
    
    >>> valeurs_annuelles_villes()
    (('Lille', 12.6, 627.1), ('Paris', 14.92, 605.4), ('Strasbourg', 12.76, 507.6),
    ('Rennes', 13.37, 757.2), ('Bordeaux', 15.32, 1156.8), ('Grenoble', 12.47, 754.5),
    ('Toulouse', 15.37, 625.1), ('Marseille', 16.6, 324.4))
    
   """
    
    # code de la fonction
    return tuple((releves[i][0], temp_moyenne(releves[i][0]), cumul_precipitations(releves[i][0])) for i in range(len(releves)))

    # --- Exercice ---
print(valeurs_annuelles_villes())
```

```python
def afficher_valeurs_annuelles_villes():
    """
    Affiche la température moyenne annuelle et le cumul annuel des précipitations
    de toutes les villes présentes dans le tuple releves
    :param: Aucun
    :return: None
    
    Le format d'affichage est :
    
    ******* Valeurs annuelles *******
    ville_1 [Température moyenne : T1 ; Cumul des précipitations : C1 mm]
    ville_2 [Température moyenne : T2 ; Cumul des précipitations : C2 mm]
    ...
    
    où ville_n correspond au nom d'une ville, Tn à sa température moyenne annuelle
    et Cn à son cumul annuel des précipitations
    
    """
    
    # code de la fonction
    print("******* Valeurs annuelles *******")
    
    for i in range(len(releves)):
        nom_ville = releves[i][0]
        # On appelle vos fonctions de calcul
        moy = temp_moyenne(nom_ville)
        prec = cumul_precipitations(nom_ville)
        
        # Affichage formaté
        print(f"{nom_ville} [Température moyenne : {moy} ; Cumul des précipitations : {prec} mm]")

    # --- Exercice ---
print(afficher_valeurs_annuelles_villes())
```

```python
import matplotlib.pyplot as plt

def graphique_temp_moyenne_et_marge():
    villes = []
    moyennes = []
    mins = []
    maxs = []

    # 1. Extraction des données
    for ville, temperatures, _ in releves:
        villes.append(ville)
        # On utilise les fonctions de base pour extraire les stats
        moyennes.append(somme(temperatures) / effectif_total(temperatures))
        mins.append(val_min(temperatures))
        maxs.append(val_max(temperatures))

    # 2. Calcul des marges (yerr attend des distances positives)
    yerr = [
        [m - mi for m, mi in zip(moyennes, mins)],  # vers le bas
        [ma - m for m, ma in zip(moyennes, maxs)]   # vers le haut
    ]

    plt.figure(figsize=(10,5))

    # Plot : un point central + des traits verticaux
    plt.errorbar(
        villes,
        moyennes,
        yerr=yerr,
        fmt='o',         # 'o' dessine un cercle pour la moyenne
        capsize=5,       # Ajoute les petites barres horizontales
        color='royalblue',
        ecolor='tomato'  # Couleur différente pour les barres d'erreur
    )

    plt.ylabel("Température (°C)")
    plt.title("Température moyenne annuelle avec marge min–max")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout() # Évite que les noms des villes soient coupés
    plt.show()

    # --- Exercice ---
graphique_temp_moyenne_et_marge()
```

```python
def ajout_donnees():
    """
    Demande à l'utilsateur de saisir une ville puis ses valeurs de températures et précipitations sur une année.
    Ajoute ensuite ces données dans le tuple releves défini en variable globale, en respectant son format de données.

    Remarque importante : 
    Le tuple releves est défini à l'extérieur de la fonction : il s'agit d'une variable globale. 
    Dans les fonctions précédentes, nous ne faisions que lire son contenu, ce qui ne posait aucun problème. Ici, nous voulons le modifier...

    Si nous ne précisons pas à la fonction que le tuple dans lequel nous voulons écrire est celui défini en variable globale, 
    elle essayera d'écrire dans une variable locale (définie dans la fonction). 
    Or, comme aucune variable locale de ce nom n'aura été définie, le programme nous retournera un message d'erreur lors de son exécution...

    Pour éviter cela, nous devons préciser à la fonction qu'elle doit utiliser la variable globale releves, 
    et ce, grâce au mot-clé global. Il faut donc ajouter au début de la fonction l'instruction : global releves


    """
    
    global releves    # utilisation de la variable globale
    
        
    # code de la fonction
    # 1. Saisie du nom de la ville
    nom_ville = input("Entrez le nom de la ville : ")

    # 2. Saisie des températures (on peut imaginer une liste de 12 mois)
    print(f"Saisie des températures pour {nom_ville} :")
    # Astuce : une petite compréhension de liste pour saisir 12 valeurs d'un coup ou une par une
    temp_list = []
    for i in range(1, 13):
        t = float(input(f"  Température mois {i} : "))
        temp_list.append(t)
        
    # On transforme la liste en tuple pour respecter le format habituel
    temperatures = tuple(temp_list)

    # 3. Saisie des précipitations
    prec_saisie = float(input("Entrez le total des précipitations annuelles : "))
    # On force le format tuple avec la virgule magique pour que les fonctions sum() fonctionnent
    precipitations = (prec_saisie,) 

    # 4. Création du nouveau relevé (format : (nom, températures, précipitations))
    nouveau_releve = (nom_ville, temperatures, precipitations)

    # 5. MISE À JOUR DU TUPLE GLOBAL
    releves = releves + (nouveau_releve,)

    print(f"Données pour {nom_ville} ajoutées avec succès !")

    # --- Exercice ---
ajout_donnees()
```

```python
def suppression_donnees():
    """
    Demande à l'utilisateur de saisir le nom d'une ville, puis supprime ses données du tuple releves.
    Si la ville entrée par l'utilisateur ne se trouve pas dans releves, ce tuple reste inchangé.
    """
    
    global releves
    
    
    # code de la fonction
    # 1. Demander le nom de la ville à supprimer
    ville_a_supprimer = input("Entrez le nom de la ville à supprimer : ")

    # 2. Reconstruire le tuple en gardant TOUT SAUF la ville saisie
    # On parcourt chaque ligne (v, t, p) et on ne la garde que si v != ville_a_supprimer
    releves = tuple(item for item in releves if item[0] != ville_a_supprimer)

    print(f"Opération terminée pour '{ville_a_supprimer}'.")

    # --- Exercice ---
suppression_donnees()
```
