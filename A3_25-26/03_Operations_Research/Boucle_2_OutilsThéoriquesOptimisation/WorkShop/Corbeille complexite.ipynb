<h4>Problème</h4>
Une entreprise de livraison en ligne nous demande de l'aide pour optimiser le remplissage de ses camions! 
En effet, avec l'augmentation du coût de transport, il est crucial de maximiser le taux de remplissage des camions pour éviter de transporter de l'air !

Nous nous intéressons à des marchandises stockées dans entrepôt A et à envoyer dans un entrepôt B. Le but est d'envoyer la totalité de la marchandise dans le plus petit nombre de camions possibles.
Nous considérerons que les marchandises à transporter sont assez légère pour ne pas se soucier du Poids Total Autorisé en Charge (PTAC), nous ne considérerons donc que le volume des marchandises comme contrainte de capacité des camions.
Nous considérerons aussi que les agents sont des excellents joueurs de tetris, et qu'ils parviennent toujours à faire rentrer les marchandises dans le camion (tant que la somme des volumes des marchandises est inférieure au volume du camion, bien sûr).

## Partie 1 : Analyse de la complexité du problème
Quel genre des problèmes pouvez vous reconnaître ici?
<blockquote>
C'est un problème d'optimisation, et plus précisément de minimisation.
</blockquote>

Quels sont les paramètres qui influent sur la taille d'une instance de ce problème ?
<blockquote>
Le nombre de camions, le nombre de colis et la capacité des camions.
</blockquote>

Nous cherchons à déterminer la classe de complexité du problème posé ici.
Nous nous intéresserons au problème de décision qui y est lié. Pouvez vous le décrire ?
<blockquote>
Étant donné une instance avec $k$ camions de capacité $C$ et $n$ colis, est-il possible de faire rentrer tous les colis dans les $k$ camions?
</blockquote>


Rappelez les principales différentes classes de complexité.
<blockquote>
P, NP, et NP-Complet.
</blockquote>

Autant vous le dire tout de suite, ce problème est NP-Complet. Quelles sont les conditions nécessaires à un problème pour qu'il soit NP-complet?
<blockquote>
Il faut que:
    <ol>
    <li> Le problème soit dans NP, c'est à dire, qu'il admette un certificat polynomial: Il existe un algorithme en temps polynomial qui est capable de verifier une solution.</li>
    <li> Le problème soit au moins aussi difficile que tous les autres problèmes de NP : On peut réduire en un temps polynomial une instance de tout autre problème NP à ce problème.</li>
</ol>

</blockquote>

Expliquez comment prouver que ces conditions sont remplies dans notre cas.
<blockquote>
Il est simple de montrer qu'une solution est réalisable. En temps polynomial, on vérifie que chaque camion ne dépasse pas sa capacité.
    
En revanche, la preuve que le problème est dans NP n'est pas demandée ici, il suffit de faire comprendre aux étudiants qu'un problème est NP-Complet si on peut y ramener tout problème de la classe NP par une réduction polynomiale.
</blockquote>


Le problème de décision associé à notre problème d'optimisation est donc NP-Complet.
Que peut on dire à propos du problème d'optimisation?
<blockquote>
Le problème d'optimisation est au moins aussi difficile, en effet, le problème d'optimisation se résume par :« Quelle est la plus petite valeur de $k$ pour laquelle la réponse est oui ? ».
</blockquote>




## Partie 2 : Solutions et complexité algorithmique du problème

Le fait qu'un problème de décision soit NP-complet veut dire qu'il n'est pas possible (à moins que P= NP) de trouver une solution en un temps polynomial en fonction de la taille de l'instance.
Pour un problème d'optimisation il n'est donc pas possible de garantir de pouvoir trouver la solution optimale en un temps polynomial.

Est-ce que toutes les solutions d'un problème de minimisation se valent ?
<blockquote>
    Non, la plupart des solutions ne sont pas optimales, l'objectif du problème d'optimisation est de trouver la solution d'un problème qui a la valeur optimale.
</blockquote>
Nous allons dans un premier temps nous intéresser au problème de trouver une solution, même sous optimale au problème.
Proposez une solution, même très mauvaise à ce problème d'optimisation, qui se calcul en temps constant.
<blockquote>
La solution la plus simple serait d'envoyer un colis par camion, c'est très rapide à calculer, quel que soit le nombre de colis. En revanche, la valeur de la solution est très élevée, nous pouvons sûrement faire mieux.
</blockquote>

Nous avons donc trouvé une solution à notre problème d'optimisation en $O(1)$, donc en temps polynomial. Alors P=NP ? Avons nous résolu un problème à un million d'euros ? Et bien non, malheureusement. Comme vous l'avez dit, toutes les solutions ne sont pas optimales. Il existe donc un très grand nombre d'algorithmes qui sont capables de trouver des solutions pour des problèmes NP-complets, si elles existent. 
Cherchons une façon de trouver des solutions, un peu plus élaborées.

Proposez une méthode (en français) pour placer les colis dans les camions, en essayant de ne pas utiliser trop de camions.
<blockquote>
Un grand nombre d'algorithmes gloutons existent et certains sont expliqués sur ce <a href='https://en.wikipedia.org/wiki/Bin_packing_problem#Single-class_algorithms'>Lien</a>. Les réponses les plus présentes seront sûrement Next-Fit (On remplis les camions un par un avec les colis jusqu'à ce qu'ils ne puissent plus rentrer, et on prend un nouveau camion, etc.) où First fit (On place les colis dans le premier camion dans lequel il rentre). 
</blockquote>

Nous allons observer le comportement de deux algorithmes gloutons. Un algorithme glouton est un algorithme qui applique une politique donnée et qui ne reviens pas sur les choix qu'il a fait au cours de son execution. Ce genre d'algorithme a le mérite d'être rapide, mais ne trouve pas toujours des solutions satisfaisantes. Dans le cas des problèmes de décision, il ne trouve même pas toujours une solution.
Est-ce que l'algorithme de placer un colis par camion est glouton ? Dans quelles conditions ne trouve-il pas de solutions au problème de décision?
<blockquote>
Oui, c'est un algorithme glouton. Il échoue à trouver des solution dès que le $n>k$, avec $n$ le nombre de colis et $k$ le nombre de camions. 
</blockquote>

Nous travaillerons sur un jeu de données généré aléatoirement. On considérera des camions avec $10m^3$ de volumes et des colis entre $1$ et $7 m^3$.

```python
import random
volume_camions = 10
Volumes_colis = [random.randrange(1,7) for i in range(100)]

def nextfit(Volumes, taille_camion):
    nb_camions = 0
    taille_restante = taille_camion
    for i in range(len(Volumes)):
        if taille_restante >= Volumes[i]:
            taille_restante = taille_restante - Volumes[i]
        else:
            nb_camions += 1
            taille_restante = taille_camion - Volumes[i]
    return nb_camions
 
def firstFit(Volumes, n, c):
     
    # Nombre de camions
    res = 0
 
    #Place restante dans les camions
    bin_rem = [0]*n
 
    # Boucle sur le nombre de colis
    for i in range(n):
         
        
        j = 0
        #Variable pour se souvenir si on a réussi à placer le colis où non
        placed = 0
 
        for j in range(res):
            if (bin_rem[j] >= Volumes[i] and placed == 0):#On cherche le premier camion dans lequel rentre le colis
                bin_rem[j] -= Volumes[i]
                placed = 1
             
        # Si le colis n'est pas placé, on crée un nouveau conteneur
        if placed == 0:
            bin_rem[res] = c - Volumes[i]
            res += 1
            

            
    return res
 
print("Volumes des colis :",Volumes_colis)
n = len(Volumes_colis)
print("Nombre de camions pour NextFit:", nextfit(Volumes_colis, volume_camions))
print("Nombre de camions pour First Fit : ",firstFit(Volumes_colis, n, volume_camions))

```

Exécutez cette cellule un certain nombre de fois. A chaque fois que vous relancez le code, une nouvelle sequence de nombre aléatoire est générée pour choisir le volume des colis. Que remarquez vous? 
<blockquote>
First-Fit à tendance à trouver des solutions avec un plus petit nombre de camions que NextFit.
</blockquote>

Pouvez vous executer mes étapes de chacun des deux algorithmes sur les 5 premiers colis de la liste? Quels sont les paramètres qui font varier le nombre d'étapes de l'algorithme? Pouvez vous, pour chacun des deux algorithmes exprimez le nombre d'étapes nécessaires en fonction de ces paramètres. Pouvez vos donner la complexité de ces algorithmes? Vous utiliserez la notation O (Prononcée Grand O).
<blockquote>
Ici, le principe est d'introduire la <a href='https://fr.wikipedia.org/wiki/Comparaison_asymptotique'>complexité asymptotique</a>. Tandis qu'il est assez évident que NextFit est linéaire en $O(n)$, avec n le nombre de colis, il apparaît que firstfit est un peu plus complexe à aborder. Vous pouvez donner l'indice aux étudiants d'étudier le pire cas (en effet, ici la complexité en moyenne nécessiterais une approche statistique et ce n'est pas le principe du prosit). Dans le pire cas, c'est à dire dans le cas où chaque colis et de la taille d'un camion (en mettant un colis par camion, on rajoute un camion a chaque itération de first fit), alors l'algorithme tourne en $O(n^2)$: Il est clair que la première boucle for tourne n fois, tandis que la deuxième itération sur le nombre de camions (donc de colis dans ce cas) déjà traités. L'opération est donc répétée une, puis deux, ..., puis $n$ fois. On a donc $\frac{n*(n+1)}{2}$ itérations.
</blockquote>

Vous venez d'être introduits au principe de la complexité asymptotique. En effet, il est assez clair qu'un algorithme linéaire tournera plus vite qu'un algorithme quadratique. Mais qu'en est-il des autres types de complexité? La différence est-elle si grande?
Le code si dessous illustre les différentes fonctions de référence pour exprimer la complexité des algorithmes. Faites varier la portée de la représentation graphique pour visualiser l'impact de la taille de l'instance sur le temps de calcul en fonction de la complexité.

```python

import numpy as np

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

from ipywidgets import interact
import ipywidgets as widgets

from ipywidgets import interact
import ipywidgets as widgets

def trace(max):
    x=np.arange(1, max, 0.1)
    
    plt.plot(x, [np.exp(x) for x in x], label="Exponentielle")
    plt.plot(x, [pow(3,x) for x in x], label="Exponentielle2")
    plt.plot(x, [pow(x, 3) for x in x], label="cubique")
    plt.plot(x, [pow(x, 2) for x in x], label="quadratique")
    plt.plot(x, [x*np.log2(x) for x in x], label="quasi-linéaire")    
    plt.plot(x, x, label="linéaire")
    plt.plot(x, [np.log2(x) for x in x], label="logarithmique")
    
    plt.xlabel("$n$", fontsize=16)
    plt.ylabel("$O(n)$", fontsize=16)
    plt.legend()
    plt.xticks(range(1, max))
    plt.gcf().set_size_inches(18, 10)
    plt.show()

interact(trace, max=widgets.Dropdown(options=[i for i in range(0,50)],min=0, max=50,step=1, value=8, description="$n$ max"));
```


Revenons à nos algorithmes gloutons. Comment pouvons nous savoir a quel point les solutions trouvées par nos algorithmes sont proches de la solution optimale?
<blockquote>
A moins que P=NP, il est impossible de prouver que ces algorithmes n'obtiennent que des solutions optimales. En revanche, il est possible que pour certaines instances du problème, ces algorithmes trouvent ces solutions optimales.

## Partie 3 : Recherche arborescente
Notre client ne se satisfait pas d'un algorithme qui ne lui permet pas d'obtenir une solution non-optimale à tous les coups.
Il a investit dans un ordinateur très puissant, et il pense que sa redoutable machine peut venir à bout de n'importe quel calcul. Il nous demande donc de trouver la solution qui minimise le nombre de camions, Eh oui, l'essence coûte cher !

Pour cela, nous décidons de résoudre le problème de décision pour un nombre de camions croissants. Dès que le problème de décision répond oui, alors on a trouvé une solution optimale.
Comment faire, étant donné un nombre de camions, pour être sûr de trouver la meilleure de toutes les solutions parmi l'espace de solution? 
<blockquote>
Il faut toutes les essayer !
</blockquote>

Comment procéderiez vous?
<blockquote>
 L'idée est d'introduire la recherche arborescente: on prend le premier colis et on le place dans le premier camion, puis on prend le deuxième colis et on le place dans le premier camion etc. Lorsque l'on arrive à une solution : c'est à dire que tous les colis sont placés dans un camion, on regarde si elle est réalisable ou non. Il faut bien insister qu'une recherche exhaustive de ce genre ne génère pas que des solutions réalisables.
    
Si la solution est réalisable, on calcul sa valeur (le nombre de camions nécessaires), puis on remonte d'une étape, on place le dernier colis dans le second camion, et on recommence: la solution est elle réalisable ? ...

 Une fois toute l'arborescence explorée, on se souvient de la meilleure solution rencontrée. C'est la solution optimale!
</blockquote>:
Considérons que nous avons deux camions, comment représenter le fait qu'un colis soit dans le premier ou le second camion? Et avec plusieurs objets? Quelle structure de données pouvons nous utiliser pour stocker l'information d'une solution? 
<blockquote>
Pour un objet, on utilise un bit.
Si il y a $n$ objets , il y a donc $n$ bits.
Pour voir toutes les solutions, il faut donc énumérer tous les nombres qu'il existe. Par exemple, sur deux bits: 00 01 10 11.
</blockquote>

Vous trouverez ci dessous un arbre avec 2 camions et 3 objets. Pouvez vous identifier les noeuds de l'arbre qui sont des solutions? Pouvez vous écrire, pour chacune des feuilles la représentation de l'information que vous avez imaginé ?
<blockquote>
Seules les feuilles de l'arbre sont des solutions.
</blockquote>

```python
import plotly.graph_objects as go

# Fonction récursive pour générer l'arbre
def dessine_arbre_camions(x, y, num_obj, nb_camions, max_objets, parent_x=None, parent_y=None, positions=[], connections=[]):
    """
    Génère un arbre où chaque niveau représente le choix d'un camion pour un objet.
    - x, y : Position actuelle du nœud
    - num_obj : Numéro de l'objet actuel
    - nb_camions : Nombre de camions possibles
    - max_objets : Nombre total d'objets
    - parent_x, parent_y : Position du parent
    - positions : Liste des positions des nœuds
    - connections : Liste des relations parent-enfant
    """
    if num_obj > max_objets:  # Arrêt quand tous les objets ont été assignés
        return
    
    # Créer les nœuds pour chaque choix de camion
    offset = 2 ** (max_objets - num_obj)  # Réduction exponentielle de l'espacement
    for i in range(1, nb_camions + 1):  # Parcourir les camions possibles
        x_pos = x + (i - (nb_camions + 1) / 2) * offset
        y_pos = y - 1  # Prochaine ligne (niveau inférieur)
        
        positions.append((x_pos, y_pos, num_obj, i))  # Ajouter le nœud avec ses détails
        if parent_x is not None and parent_y is not None:
            connections.append(((x_pos, y_pos), (parent_x, parent_y)))  # Relier au parent
        
        # Appel récursif pour le niveau suivant
        dessine_arbre_camions(x_pos, y_pos, num_obj + 1, nb_camions, max_objets, x_pos, y_pos, positions, connections)

# Initialisation des paramètres
n = 3  # Nombre d'objets
k = 2  # Nombre de camions

positions = []
connections = []
# Ajouter la racine vide
racine_x, racine_y = 0, 0
positions.append((racine_x, racine_y, 0, 0))  # La racine est ajoutée en premier
# Relier la racine au premier niveau
dessine_arbre_camions(racine_x, racine_y, 1, k, n, racine_x, racine_y, positions, connections)

# Séparer les coordonnées et les informations des nœuds
x_coords, y_coords, objets, camions = zip(*positions)

# Tracer les connexions (arêtes)
edges_x = []
edges_y = []
for (x_start, y_start), (x_end, y_end) in connections:
    edges_x += [x_start, x_end, None]  # None pour segmenter les lignes
    edges_y += [y_start, y_end, None]

# Créer une figure avec Plotly
fig = go.Figure()

# Ajouter les arêtes
fig.add_trace(go.Scatter(
    x=edges_x,
    y=edges_y,
    mode='lines',
    line=dict(color='black', width=1),
    hoverinfo='none'  # Pas de hover sur les lignes
))

# Ajouter les nœuds
fig.add_trace(go.Scatter(
    x=x_coords,
    y=y_coords,
    mode='markers+text',
    marker=dict(size=10, color=['blue' if obj == 0 else 'red' for obj in objets]),
    text=[f"Racine" if obj == 0 else f"Objet {obj}, Camion {cam}" for obj, cam in zip(objets, camions)],
    textposition="top center",
    hoverinfo='text'  # Affiche uniquement le texte au survol
))

# Ajuster les axes et le design
fig.update_layout(
    title="Arbre des Objets et Camions (Plotly)",
    showlegend=False,
    xaxis=dict(showline=False, zeroline=False, showgrid=False, visible=False),
    yaxis=dict(showline=False, zeroline=False, showgrid=False, visible=False),
    xaxis_range=[-2**n, 2**n],
    yaxis_range=[-n-1, 1],
    plot_bgcolor='white'
)

# Afficher le graphique
fig.show()

```

Pouvez vous compter la taille de l'arbre en fonction de n, le nombre de colis et k le nombre de camions? De quelle nature est cet algorithme, qui explore chaque noeud de l'arbre?
<blockquote>
C'est un algorithme exponentiel, en $O(k^n)$, la taille de l'arbre.
</blockquote>


La représentation que vous utilisez ne fonctionne qu'avec deux camions, comment faire si vous utilisez plus de deux camions ?
<blockquote>
Il faut utiliser la notation en base $k$, pour $k$ camions et appliquer le même algorithme: on compte de $1$ à $k^n$, et on regarde les feuilles.
</blockquote>

Le code ci dessous recherche toutes les façons de placer $n$ colis parmi $k$ camions de capacité $C$. Faites varier les paramètres pour observer la vitesse de calcul des différents jeux de données.

```python
import random 
colis = [random.randrange(2, 5) for i in range(10)] #Generation des n objets 
taille_camion = 10 #capacité C des camions
nb_camions = 3 #k camions
#Executez cette cellule quand vous voulez changer de jeu de donnée
```

```python
def numberToBase(n, b,nb_colis):
    digits = [0]*len(colis)
    if n == 0:
        return digits
    i = 1
    while n:
        digits[-i] = int(n % b)
        n //= b
        i+=1
    return digits



def calcul_solution(number):
    affectations = numberToBase(number,nb_camions,len(colis))
    charge_camions = [0]*nb_camions
    for i in range(len(colis)):
        charge_camions[affectations[i]] += colis[i]
    
    return charge_camions, nb_camions-charge_camions.count(0)
nombre_solutions = 0
solution_optimale = nb_camions
for i in range(nb_camions**len(colis)):
    charges, nb= calcul_solution(i)
    if all(charges[j]<=taille_camion for j in range(len(charges))): # La solution est réalisable
        nombre_solutions+= 1 
        if nb < solution_optimale:
            solution_optimale = nb

print("Il y a ",nombre_solutions," solutions réalisables, et la solution optimale est de ",solution_optimale," camions")
```

Vous avez pu identifier la classe de complexité du problème de décision d'un problème classique du domaine. 
Il est important de bien faire la différence entre un problème de décision et d'optimisation : un problème de décision NP-complet peut avoir un problème d'optimisation d'une classe de complexité supérieure.

Il est aussi important de bien faire la distinction entre complexité d'un problème, et complexité d'un algorithme. 
Pour un problème de décision, il n'existe pas d'algorithmes polynomiaux qui soient capables de donner une solution, à moins que P = NP. En revanche, pour un problème d'optimisation, certains algorithmes polynomiaux trouveront des solutions réalisables, mais sans garanties d'optimalités.

Enfin, vous avez pu voir la vitesse à laquelle croit le temps de calcul pour un algorithme exponentiel. C'est ici que réside toute la complexité de l'informatique. Même pour de petites instance, il est très difficile de résoudre optimalement un grand nombre de problèmes classiques.
